```python
from sklearn import tree
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
```

```python
wine = load_wine()
```

```python
wine
```

```python
wine.data
```

```python
wine.target
```

```python
import pandas as pd
pd.concat([pd.DataFrame(wine.data),pd.DataFrame(wine.target)],axis=1)
```

```python
wine.feature_names
```

```python
wine.target_names
```

```python
Xtrain,Xtest, Ytrain, Ytest = train_test_split(wine.data,wine.target,test_size=0.3)
```

```python
Xtrain.shape
```

```python
wine.data.shape
```

```python
clf = tree.DecisionTreeClassifier(criterion="gini")
clf = clf.fit (Xtrain, Ytrain)
score = clf.score(Xtest,Ytest)
score
```

```python
clf = tree.DecisionTreeClassifier(criterion="entropy")
clf = clf.fit (Xtrain, Ytrain)
score = clf.score(Xtest,Ytest)
score
```

```python
import graphviz
feature_name = ['A','B','C','D','E','F','G','H','I','J','K','L','M']
dot_data = tree.export_graphviz(clf
                              ,feature_names=feature_name
                              ,class_names=["Gin", "Rum", "Whiskey"]
                              ,filled = True
                              ,rounded = True
                               )
graph = graphviz.Source(dot_data)
graph
```

```python
clf.feature_importances_
#特征的重要性
[*zip(feature_name,clf.feature_importances_)]
```

```python
clf = tree.DecisionTreeClassifier(criterion="entropy",random_state=30)
#每个节点随机选取不同特征
clf = clf.fit (Xtrain, Ytrain)
score = clf.score(Xtest,Ytest)
score
```

```python
clf = tree.DecisionTreeClassifier(criterion="entropy"
                                  ,random_state=30
                                  ,splitter="random"
                                  # default splitter "best"
                                 )
clf = clf.fit (Xtrain, Ytrain)
score = clf.score(Xtest,Ytest)
score
```

```python
dot_data = tree.export_graphviz(clf
                              ,feature_names=feature_name
                              ,class_names=["Gin", "Rum", "Whiskey"]
                              ,filled = True
                              ,rounded = True
                               )
graph = graphviz.Source(dot_data)
graph
```

```python
clf.feature_importances_
[*zip(feature_name,clf.feature_importances_)]
```

```python
#测试训练集的拟合程度
score_train = clf.score(Xtrain,Ytrain)
score_train
```

```python
#减枝策略
#3个减枝策略可以灵活应用
clf = tree.DecisionTreeClassifier(criterion="entropy"
                                  ,random_state = 30
                                  ,splitter = "random"
                                  ,max_depth = 3
                                  ,min_samples_leaf = 10
                                  ,min_samples_split = 10
                                  )
clf = clf.fit (Xtrain, Ytrain)
dot_data = tree.export_graphviz(clf
                              ,feature_names=feature_name
                              ,class_names=["Gin", "Rum", "Whiskey"]
                              ,filled = True
                              ,rounded = True
                               )
graph = graphviz.Source(dot_data)
graph
```

```python
score_train = clf.score(Xtrain,Ytrain)
score_train
```

```python
score = clf.score(Xtest,Ytest)
score
```

```python
#确定最优减枝参数
import matplotlib.pyplot as plt

test=[]
for i in range(10):
    clf = tree.DecisionTreeClassifier(max_depth=i+1
                                      ,criterion = "entropy"
                                      ,random_state = 30
                                     )
    clf = clf.fit (Xtrain, Ytrain)
    score = clf.score(Xtest,Ytest)
    test.append(score)
    
plt.plot(range(1,11),test,color="red",label="max_depth")
plt.legend()
plt.show()
```

```python
test=[]
for i in range(100):
    clf = tree.DecisionTreeClassifier(min_samples_leaf = i+1
                                      ,criterion = "entropy"
                                      ,random_state = 30
                                     )
    clf = clf.fit (Xtrain, Ytrain)
    score = clf.score(Xtest,Ytest)
    test.append(score)
    
plt.plot(range(1,101),test,color="red",label="min_samples_leaf")
plt.legend()
plt.show()
```

```python
test=[]
for i in range(100):
    clf = tree.DecisionTreeClassifier(min_samples_split = i+1
                                      ,criterion = "entropy"
                                      ,random_state = 30
                                     )
    clf = clf.fit (Xtrain, Ytrain)
    score = clf.score(Xtest,Ytest)
    test.append(score)
    
plt.plot(range(1,101),test,color="red",label="min_samples_split")
plt.legend()
plt.show()
```
